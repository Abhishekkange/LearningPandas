#Plotting Graphs


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.style.use('ggplot')
```

```python
#scatterplot
sns.scatterplot(x="area",y="price",data=dataset)
```

```python
#scatter plot with hue
sns.scatterplot(x="area",y="price",data=dataset,hue="furnishingstatus")
```

# **Historgram**
Histogram - On the X axis we hav the class intervals and on y axis we have the count/frequency.

It is used to summarize the data. For eg in below graph we hav class interval as area and corresponding to it we have the no. of houses lying between two area points.

```python
#Histogram
sns.histplot(dataset['area'])
```

```python
#Adding Bins in graph, It means Number of Columns in graph
sns.histplot(dataset['area'],bins=3)
```

# **Bar Graph**
Bar graph - We have some categorical data on the x axis and corresponding to that data we can have the frequence of that particular data on the y axis.


```python
sns.barplot(x="furnishingstatus",y="price",data=dataset)
```

```python
#Adding Hue
sns.barplot(x="furnishingstatus",y="price",data=dataset,hue="airconditioning")
```

# **DisPlot**
Displot stands for Distribution Plot

```python
sns.displot(x="area",y="price",data=dataset,kind="kde")
```

# **Pie Chart**

```python
import matplotlib.pyplot as plt
```

```python
status_count = dataset['furnishingstatus'].value_counts()
slabels = dataset['furnishingstatus'].value_counts().index
plt.pie(status_count,labels=slabels)
```

```python
#add percentage
status_count = dataset['furnishingstatus'].value_counts()
slabels = dataset['furnishingstatus'].value_counts().index
plt.pie(status_count,labels=slabels,autopct='%.2f%%', startangle=90)
plt.title('Furnishing Status Chart')
plt.show()
```
